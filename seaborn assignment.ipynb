```python
'''Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.'''
#importing seaborn library as sns
import seaborn as sns
#loading dataset
fmri = sns.load_dataset("fmri")
```

```python
fmri
```

```python
#lineplot
sns.lineplot(x=fmri.timepoint,y=fmri.signal)
```

```python
'''Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.'''

#importing matplot for adjusting size
import matplotlib.pyplot as plt
#loading titanic data
titanic = sns.load_dataset("titanic")
```

```python
#taking first five data using head()
new = titanic.head()
```

```python
new
```

```python
#plotting box 
sns.boxplot(x=new.pclass,y=new.age,data = new,hue ='fare')
plt.figure(figsize=(1,2))
```
